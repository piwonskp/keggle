```python
def hist_by_private(df, x):
  grid = seaborn.FacetGrid(df, hue="Private", palette="pastel")
  return grid.map(plt.hist, x, bins=20, alpha=0.7)
  
hist_by_private(df, "Outstate")
hist_by_private(df, "Grad.Rate")
```

```python
from IPython.display import display

df.loc[df['Grad.Rate'] > 100, 'Grad.Rate'] = 100
display(df[df['Grad.Rate'] > 100])
hist_by_private(df, 'Grad.Rate')
```

```python
from sklearn.cluster import KMeans
from sklearn.model_selection import train_test_split
from IPython.display import display

Y_COL = 'Private'
y = df[Y_COL]
x = df[df.columns.difference([Y_COL, 'Unnamed: 0'])]
x_train, x_test, _, y_test = train_test_split(x, y)

kmeans = KMeans(2).fit(x_train)
kmeans.cluster_centers_
```

```python
cluster = y_test.map(lambda x: 1 if x == "Yes" else 0)
cluster.name = 'Cluster'

pandas.concat([y_test, x_test, cluster], axis=1)
```

```python
from sklearn.metrics import confusion_matrix, classification_report


y_pred = kmeans.predict(x_test)
print(confusion_matrix(cluster, y_pred))
print(classification_report(cluster, y_pred))
```
